```python
# Boolean indexing(True/False)
names = np.array(["Seoul", "Daejun", "chungju", "seoul", "chungju", "Daejun", "Daejun"])
data = np.random.randn(7,4) # 7 X 4
print(data)
print(names == "Seoul") # 1, 0, 0, 0, 0, 0, 0
print("------------------------------------------------")
print(data[names == "Seoul"])
print(data[names == "Seoul", 2:])
print(data[names == "Seoul", 3])
print(names != "Seoul")
print(~(names != "Seoul"))
mask = (names == "Seoul") | (names == "chungju")
print(mask)
print(data[mask])
```

```python
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, True, False])
print(zip(xarr, yarr, cond)) # zip : 요소별로 같은 열의 데이터를 묶음
print(list(zip(xarr, yarr, cond)))

result = [(x if c else y) for x,y,c in zip(xarr, yarr, cond)]
print("result = ", result)

# 위의 result를 함수화한 것이 np.where(cond, TRUE, FALSE)
result = np.where(cond, xarr, yarr) # cond가 참이면 X, 아니면 Y
print("result = ", result)

arr = np.random.randn(4,4)
print(arr)
print(np.where(arr>0, 2, -2))
```

```python
# 문제. 다음 데이터를 이용하여 두개 다 참일 때는 0, cond1이 참일 때는 1,
# cond2가 참일 때는 2, 그 외일때는 3으로 처리하여 출력하도록 하시요.
# 조건문은 4개, 삼항연산자는 3개
cond1 = np.array([1,0,1,1,0,0,1], dtype = bool)
cond2 = np.array([0,1,1,1,0,1,1], dtype = bool)
result1 = np.where(cond1, np.where(cond2, 0, 1), np.where(cond2, 2, 3))
result2 = np.where(cond1 & cond2, 0, np.where(~cond1 & ~cond2, 3, np.where(cond1, 1, 2)))
result3 = np.where(cond1 & cond2, 0, np.where(cond1, 1, np.where(cond2, 2, 3)))
result4 = 1 * (cond1 & ~cond2) + 2 *(cond2 & ~cond1) + 3 * ~(cond1 | cond2)
print(result1)
print(result2)
print(result3)
print(result4)
```

```python
# 행렬의 종류
# - 정방행렬 : 정사각형(행과 열의 수가 같은 행렬)
# - 대칭행렬 : 우상단과 좌하단이 같은 행렬
# - 대각행렬 : scale을 변화시키는 요소
# - 단위행렬(indentity) : 행렬곱의 항등원(행렬*역행렬 = 단위행렬)
# - 역행렬 : 행렬의 나눗셈을 위해 곱해지는 행렬(정방행렬만 가능)
# - 특이행렬 : 역행렬이 없는 행렬...
# - 직교행렬 : 축의 정직교, 각 축간의 내적이 0인 행렬
# - 공분산행렬, 상관계수행렬, 거리행렬 : 정방행렬이면서 대칭행렬
# - 공분산 : 두 수 간의 관계. ((하나수-평균) * (다른수-평균)) / (n-1)
# - 상관계수 : (표준화 -1~1) : 공분산 / 표준편차의 곱
```

```python
import numpy as np
F1 = np.array([[1,2], [3,4]])
F2 = np.array([[2,2], [2,2]])
print(F1)
print(F2)
print(F1 + F2)
print(F1 * F2)
print(F1 @ F2) # 행렬 곱도 내적 - 두 벡터의 사이각을 구해짐
               # 2개의 특성이 만들어짐
               # 2X2 * 2X3 -> 차원 확장
# 한꺼번에 차원축소를 하면 버려지는 차원이 많으면 손실이 큼
# 그래서 조금씩 여러번 축소할 필요가 있다 => deep learning
```

```python
# 방정식의 해
# 2 x1 + x2 -2 x3 = -3
# 3 x1 +       x3 = 5
#   x1 + x2 -  x3 = -2
A = np.array([[2,1,-2], [3,0,1], [1,1,-1]])
print(A)
b = np.transpose(np.array([[-3,5,-2]])) # 열로 전치
print(b)

# np.linalg.solve: 역행렬을 내부적으로 구해서 행렬곱
# np.linalg.inv: 역행렬, np.dot, @: 행렬곱
x = np.linalg.solve(A, b) # A의 역행렬을 구해서 b와 곱해줌
print(x)
print(np.dot(np.linalg.inv(A), b)) # 행렬곱
print(np.linalg.inv(A) @ b) # 행렬은 나눗셈이 없으므로 역행렬을 곱함

# 검산 [1, -1, 2]
# 2 x1 + x2 -2 x3 = -3
# 3 x1 +       x3 =  5
#   x1 + x2 -  x3 = -2
print("검산결과 2*1 +(-1) -2 *2 =", 2*1 +(-1) -2 *2)
```

```python
a = np.array([0,1,0])  # y축
b = np.array([1,0,0])  # x축
res = np.dot(a,b) # 내적 사이각 (내적=0 => 직각)
print(res)

print(np.linalg.norm(a)) # norm: 벡터의 크기
print(np.linalg.norm(b))
print(np.linalg.norm(a) * np.linalg.norm(b))
rad = res / (np.linalg.norm(a) * np.linalg.norm(b)) # =cos theta(radian)
print("cos theta", rad)
print(np.arccos(rad)) # =theta
print(np.degrees(np.arccos(rad))) # 각도
```
